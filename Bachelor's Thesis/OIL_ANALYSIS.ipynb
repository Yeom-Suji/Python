```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 프레임 생성
data_production = {
    'DATE': ['2000-01-01', '2001-01-01', '2002-01-01', '2003-01-01', '2004-01-01', '2005-01-01', '2006-01-01', '2007-01-01', 
             '2008-01-01', '2009-01-01', '2010-01-01', '2011-01-01', '2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01', 
             '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01', '2021-01-01', '2022-01-01', '2023-01-01', '2024-01-01', '2025-01-01'],
    'Crude_Oil_Production': [3661000, 3571700, 3248200, 3754000, 3990000, 4014000, 4040000, 4094000, 3942916.667, 3725000, 
                             3688000, 3621000, 2800000, 2709083.333, 2764500, 2920916.667, 3675333.333, 3814750, 3468888.889, 
                             2193000, 2024333.333, 2482500, 2564583.333, 3117986.458, 3194986.458, 3267751.458]
}

data_exports = {
    'DATE': ['2000-01-01', '2001-01-01', '2002-01-01', '2003-01-01', '2004-01-01', '2005-01-01', '2006-01-01', '2007-01-01', 
             '2008-01-01', '2009-01-01', '2010-01-01', '2011-01-01', '2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01', 
             '2016-01-01', '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01', '2021-01-01', '2022-01-01', '2023-01-01', '2024-01-01', '2025-01-01'],
    'Crude_Oil_Exports': [2096687, 1925032, 1982782, 2262897, 2279872, 2185512, 2256663, 2225072, 2131074, 2019443, 
                          2102427, 2185425, 1692317, 1069966, 991951, 1162162, 1775292, 1850538, 1394993, 530525, 
                          444667.5, 717525, 896906.2, 1400000, 1470000, 1536150]
}

df_production = pd.DataFrame(data_production)
df_exports = pd.DataFrame(data_exports)
df_production['DATE'] = pd.to_datetime(df_production['DATE'])
df_exports['DATE'] = pd.to_datetime(df_exports['DATE'])

# 데이터 병합
df = pd.merge(df_production, df_exports, on='DATE')
df['Year'] = df['DATE'].dt.year

# 선 그래프
plt.figure(figsize=(14, 7))
sns.lineplot(x='DATE', y='Crude_Oil_Production', data=df, marker='o', label='Crude Oil Production', color = 'g')
sns.lineplot(x='DATE', y='Crude_Oil_Exports', data=df, marker='o', label='Crude Oil Exports', color = 'orange')

plt.title('Crude Oil Production and Exports Over Time')
plt.xlabel('Year')
plt.ylabel('Barrels per day')
plt.grid(True)
plt.xticks(df['DATE'], df['Year'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
