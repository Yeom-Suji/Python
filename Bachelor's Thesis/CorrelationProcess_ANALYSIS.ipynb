## 이란 GDP만 남기기(1993-2021)

```python
import numpy as np
import pandas as pd
```

```python
world_gdp = pd.read_csv(r"C:\Users\LG\Desktop\Paper\Paper_Code\DATA\API_WORLDBANK_GDP_raw.csv", encoding = 'utf-8')
world_gdp
```

```python
# 이란 gdp만 남김
iran_gdp = world_gdp[world_gdp['Country Name'].isin(["Iran, Islamic Rep."])]
iran_gdp
```

```python
missing_values = iran_gdp.isna().any()

# 결측치 어디있나 확인
print("결측치 확인:")
print(missing_values[missing_values].index.tolist())
```

```python
iran_gdp.columns
```

```python
# '1991', '1992'을 포함한 이전 년도 데이터 열 모두 삭제
iran_gdp = iran_gdp.drop(['Country Code', 'Indicator Name', 'Indicator Code',
    '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
    '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
    '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
    '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996','1997','1998','1999','2022'], axis=1)

iran_gdp = iran_gdp.reset_index(drop=True)
iran_gdp
```

```python
# 칼럼 이름 변수로 변경
iran_gdp.rename(columns={'Country Name': 'YEAR'}, inplace=True)
iran_gdp
```

```python
# "Iran, Islamic Rep." 값을 "GDP"로 변경
iran_gdp['YEAR'] = iran_gdp['YEAR'].replace('Iran, Islamic Rep.', 'GDP')
iran_gdp

```

```python
# 데이터프레임 전치
iran_gdp = iran_gdp.transpose()

# 전치된 데이터프레임의 첫 번째 행을 열 이름으로 설정
iran_gdp.columns = iran_gdp.iloc[0]
iran_gdp = iran_gdp.drop(iran_gdp.index[0])
```

```python
iran_gdp = iran_gdp.reset_index()
iran_gdp
```
